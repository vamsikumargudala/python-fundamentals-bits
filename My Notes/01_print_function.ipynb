## Basic Usage

````python
print("Hello World")
````

````python
# Single or double quotes - both work
print('Single quotes')
print("Double quotes")
````

## Strings vs Expressions

- Text in quotes → printed as-is
- No quotes → evaluated first

````python
print('3 + 4')   # Output: 3 + 4 (string)
print(3 + 4)     # Output: 7 (evaluated)
````

## Multiple Arguments

Pass multiple values separated by commas. They're joined with spaces by default.

````python
print('Name:', 'Vamsi', 'Age:', 25)
````

````python
# Mix strings and expressions
print('Result:', 10 * 5)
````

## The `sep` Parameter

Changes the separator between arguments (default is space).

````python
print('a', 'b', 'c')              # a b c
print('a', 'b', 'c', sep='-')     # a-b-c
print('a', 'b', 'c', sep='')      # abc
print('a', 'b', 'c', sep='\n')    # each on new line
````

## The `end` Parameter

Changes what's printed at the end (default is newline `\n`).

````python
print('First', end=' ')
print('Second')           # First Second (same line)
````

````python
print('Loading', end='...')
print('Done!')            # Loading...Done!
````

## Escape Characters

| Code | Meaning |
|------|--------|
| `\n` | New line |
| `\t` | Tab |
| `\\` | Backslash |
| `\'` | Single quote |
| `\"` | Double quote |

````python
print('Line 1\nLine 2')
print('Column1\tColumn2')
````

## Quick Reference

```python
print(value)                    # Basic
print(a, b, c)                  # Multiple values
print(a, b, sep='-')            # Custom separator
print(a, end='')                # No newline at end
print(a, b, sep='-', end='!')   # Both
```
